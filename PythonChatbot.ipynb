```python
def calculator():
    # 1. Get user input for numbers and operator
    try:
     while True:
        print("-----------------------------------------------------------")
        print("Type 'exit' to exit and 'con' to continue")
        ask=input().lower()
        if ask == "exit":
          print("-----------------------------------------------------------")
          chatbot_menu()
          print("-----------------------------------------------------------")
        elif ask=="con":
          print("-----------------------------------------------------------")
          num1 = float(input("Enter first number: "))
          operator = input("Enter operator (+, -, *, /): ")
          num2 = float(input("Enter second number: "))

          # 2. Perform calculation based on operator
          if operator == '+':
              result = num1 + num2
          elif operator == '-':
              result = num1 - num2
          elif operator == '*':
              result = num1 * num2
          elif result == 67.0:
              print("six seven 😝")
          elif operator == '/':
              # Check for division by zero
              result = num1 / num2 if num2 != 0 else "Error: Division by zero"
          else:
              result = "Invalid operator"

          print(f"Result: {result}")
        else:
          print("Invalid Input")


    except ValueError:
        print("Invalid input. Please enter numeric values.")
```

```python
def unit_converter():
 while True:
  print("-------------------------")
  print("1.Kilometer to meter")
  print("2.Inches to feet")
  print("3.Farenheight to Celcius")
  print("4.Celcius to farenheight")
  print("5.Hours to seconds")
  print("6.Seconds to hours")
  print("7.Inches to centimeter")
  print("8.exit")
  print("----------------------------")
  ask=int(input( "Enter the number of what would you like to do: "))
  print("--------------------------------------------------------------")
  if ask==1:
    kilo_meter()
    print("-----------------------------------------------------------")
  if ask==2:
    inch_feet()
    print("-----------------------------------------------------------")
  if ask==3:
    faren_cel()
    print("-----------------------------------------------------------")
  if ask ==4:
    cel_faren()
    print("-----------------------------------------------------------")
  if ask==5:
    hour_sec()
    print("-----------------------------------------------------------")
  if ask==6:
    sec_hour()
    print("-----------------------------------------------------------")
  if ask ==7:
    inch_centi()
    print("-----------------------------------------------------------")
  if ask==8:
    print("-----------------------------------------------------------")
    chatbot_menu()
    print("-----------------------------------------------------------")
def kilo_meter():
  ch=int(input("Enter the number in kilometers: "))
  formula=ch*1000
  print(round(formula,1),"meters")
def inch_feet():
  ch=int(input("Enter the number in inches: "))
  formula=ch/12
  print(round(formula,1),"feet")
def faren_cel():
  # C = (F - 32) x 5/9
  # F = (C x 9/5) + 32
  ch=int(input("Enter the number in farenheight: "))
  formula= (ch - 32) * 5/9
  print(round(formula,1),"Celcius")
def cel_faren():
  ch=int(input("Enter the number in celcius: "))
  formula=(ch * 9/5) + 32
  print(round(formula,1),"Farenheight")
def hour_sec():
  ch=int(input("Enter the number in hours: "))
  formula=(ch*60) * 60
  print(round(formula,1),"Seconds")
def sec_hour():
  ch=int(input("Enter the number in seconds: "))
  formula=ch/3600
  print(round(formula,1),"Hours")
def inch_centi():
  ch=int(input("Enter the number in inches: "))
  formula=ch*2.54
  print(round(formula,2),"Centimers")


```

```python
def quiz():
  score=0
  questions=["1.How many zeros does a googol have? ",
             "2.What is the pronoun in this sentence; Her pen is on the desk.",
             "3.How many continents are in the world? ",
             "4.True or False: Plants are living things.",
             "5.Which symbol is used for a comment in python? "

  ]
  answer_ch=[
      ["a. 20","b. 50","c. 100","d. 80"],
      ["a. Her","b. desk","c. on","d. pen"],
      ["a. 3","b. 1","c. 7","d. 50"],
      ["a. True","b. False","c. All of them","d. None of them"],
      ["a. $","b. //","c. '' ","d. #"]
             ]
  answers=[ "c","a","c","a","d"]

  for q in range(0,len(questions)):
    print(questions[q])
    for a_ch in range(0,len(answer_ch[q])):
      print(answer_ch[q][a_ch])
    ask=input("What is the answer: ")
    if ask==answers[q]:
      score+=1
      print("Correct!")
      print("Your score is now:",score)
    else:
      print("Incorrect, the answer was,",answers[q])
  print("Your total score is",score)
  print("-----------------------------------------------------------")
  if score==5:
    print("Excellent!")
  if score==4:
    print("Good")
  if score==3:
    print("There's always room for improvement!")
  if score<=2:
    print("You need to study more.Better luck next time")
```
